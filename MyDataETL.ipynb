and might be all features that I've explored so far. But I can also transform or create new features from the ones that I have, specially considering the analysis that I wanna do. Base on that, I'll create a feature for year, month and hoursPlayed, just to make my life easier when I will be analysing it.

```python
def get_datetime(datetime_obj: int, value: str) -> datetime:
    ''' Just made a function to get the year and month based on the endTime ''' 
    
    if value == 'year':
        return datetime.strptime(str(datetime_obj), "%Y-%m-%d %H:%M").year
    
    return datetime.strptime(str(datetime_obj), "%Y-%m-%d %H:%M").month
```

```python
streaming['year'] = streaming['endTime'].apply(lambda x: get_datetime(x, 'year'))
```

```python
streaming['month'] = streaming['endTime'].apply(lambda x: get_datetime(x, 'month'))
```

```python
streaming['hoursPlayed'] = streaming['msPlayed'].transform(lambda x: x*1e-3/3600)
```
